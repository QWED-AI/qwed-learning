## 3. Your First Verification

Let's verify a simple math problem:

````python
from qwed_new.core import DiagnosticStatus
result = client.verify_math("What is the derivative of x^2?")

print(f"Status: {result.status.value}")
print(f"Answer: {result.developer_fields.get('value')}")
print(f"Proof Ref: {result.proof_ref}")
print(f"Method: {result.developer_fields.get('method')}")
````

**Expected Output:**
```
Status: VERIFIED
Answer: 2*x
Proof Ref: sha256:...
Method: symbolic
```

## 4. Understanding the Result

The `DiagnosticResult` object contains three layers:

````python
# Inspect the full result
print("=" * 50)
print("Full DiagnosticResult")
print("=" * 50)
print(f"status: {result.status.value}")
print(f"agent_message: {result.agent_message}")
print(f"proof_ref: {result.proof_ref}")
print(f"is_verified: {result.is_verified}")
print(f"is_authoritative: {result.is_authoritative}")
print(f"\ndeveloper_fields: {result.developer_fields}")
````

## 5. Try More Verifications

### Calculus

````python
# Integration
result = client.verify_math("Integrate x^2 from 0 to 1")
print(f"∫x² dx from 0 to 1 = {result.developer_fields.get('value')}")
````

### Algebra

````python
# Solve equation
result = client.verify_math("Solve x^2 - 4 = 0")
print(f"Solutions: {result.developer_fields.get('value')}")
````

### Finance

````python
# Compound interest
result = client.verify_math(
    "Calculate compound interest on $10,000 at 5% for 10 years"
)
print(f"Final amount: ${result.developer_fields.get('value'):,.2f}")
````

##  6. What Happens When Verification Fails?

````python
# Try an invalid query
result = client.verify_math("What is the color of math?")

if not result.is_verified:
    print(f"❌ Verification failed!")
    print(f"Status: {result.status.value}")
    print(f"Message: {result.agent_message}")
else:
    print(f"✅ Verified: {result.developer_fields.get('value')}")
````
